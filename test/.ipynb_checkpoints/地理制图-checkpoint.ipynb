## 蒙古国制图

```python
# 导入cartopy和matplotlib等模块
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.io.img_tiles as cimgt
import cartopy.io.shapereader as shpreader
import matplotlib.pyplot as plt
```

### 地形绘制

```python
# 创建一个Stamen地形网络地图对象，设置缩放级别为8
stamen_terrain = cimgt.Stamen('terrain-background')
zoom_level = 4
```

```python
# 创建一个地形特征，使用50m分辨率的Natural Earth数据
terrain = cfeature.NaturalEarthFeature('physical', 'populated_places', '50m',
                                       edgecolor='face',
                                       facecolor=cfeature.COLORS['land'])

# 创建一个地图投影，使用PlateCarree（等距圆柱投影）
proj = ccrs.PlateCarree()

# 创建一个GeoAxes对象，使用proj作为投影
ax = plt.axes(projection=proj)

# 添加地形特征到地图上，使用terrain.crs作为特征的投影
ax.add_feature(terrain, facecolor='gray')

# 设置地图的范围，使用德国的经纬度边界
ax.set_extent([5.5, 15.5, 47, 55], crs=proj)

# 显示地图
plt.show()
```

```python
# 创建一个等距圆柱投影的地图
fig = plt.figure(figsize=(16,8))
ax = fig.add_axes(111,projection=ccrs.PlateCarree())

# 设置地图的显示范围为蒙古国的经纬度范围
ax.set_extent([87, 120, 40, 52])

# 添加海岸线、国界线、河流、湖泊等特征
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, facecolor='none',linestyle='--',linewidth=0.8)
ax.add_feature(cfeature.RIVERS)
ax.add_feature(cfeature.LAKES)

# 添加行政区划特征
ax.add_feature(cfeature.STATES,alpha=0.4,linestyle='--',linewidth=0.3)

# 添加网络地图到地图上，设置透明度为0.5
ax.add_image(stamen_terrain, zoom_level, alpha=0.5)

# 添加网格线和刻度标签
ax.gridlines(draw_labels=True,linewidth=0.8,color='k',alpha=0.3,linestyle='--')

# 添加标题
plt.title('Topography and Provinces of Mongolia')

# 显示地图
plt.show()
```

## OSM绘制

```python
# 导入cartopy和matplotlib等模块
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.io.img_tiles as cimgt
import cartopy.io.shapereader as shpreader
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
```

```python
# 创建一个openstreetmap对象
osm = cimgt.OSM()
```

```python
# 创建一个等距圆柱投影的地图
fig = plt.figure(figsize=(16,8))
ax = fig.add_axes(111,projection=ccrs.PlateCarree())

# 设置地图的显示范围为蒙古国的经纬度范围
ax.set_extent([87, 120, 40, 52])

# 添加海岸线、国界线、河流、湖泊等特征
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, facecolor='none',linestyle='--',linewidth=0.8)
ax.add_feature(cfeature.RIVERS)
ax.add_feature(cfeature.LAKES)

# 添加行政区划特征
ax.add_feature(cfeature.STATES,alpha=0.4,linestyle='--',linewidth=0.3)

# 添加网络地图到地图上，设置透明度为0.5
ax.add_image(stamen_terrain, zoom_level, alpha=0.5)

# 添加openstreetmap的底图到地图上，使用osm.crs作为底图的投影
ax.add_image(osm, 6)  # 8是缩放级别，越大越详细

# 添加网格线和刻度标签
ax.gridlines(draw_labels=True,linewidth=0.8,color='k',alpha=0.3,linestyle='--')

# 添加标题
plt.title('Topography and Provinces of Mongolia')

# 显示地图
plt.show()
```

## 中国制图

```python
fig = plt.figure(figsize=(16,9))
res = ['110m', '50m', '10m']
extents = [75, 150, 15, 60]

proj = ccrs.PlateCarree()
for i, res in enumerate(['110m', '50m', '10m']):
    ax = fig.add_subplot(1, 3, i+1, projection=proj)
    ax.set_extent(extents, crs=proj)

    ax.add_feature(cfeature.OCEAN.with_scale(res))
    ax.add_feature(cfeature.LAND.with_scale(res), edgecolor='black')
    ax.add_feature(cfeature.LAKES.with_scale(res), edgecolor='black')
    ax.add_feature(cfeature.RIVERS.with_scale(res))
    ax.add_feature(cfeature.BORDERS.with_scale(res))
    ax.gridlines(linestyle='--')

    ax.set_title('resolution=' + res)

plt.show()
```

## 粤港澳大湾区制图

```python
import geopandas as gpd

# 导入模块
import cartopy
import cartopy.crs as ccrs
import cartopy.io.img_tiles as cimgt
import cartopy.io.shapereader as shpreader
import matplotlib.pyplot as plt
from adjustText import adjust_text

plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
```

### 文件读取

```python
# 读取文件
df = gpd.read_file(r"E:\Netease\工程数据\粤港澳大湾区.shp")
df.head()
```

```python
df
```

```python
# 筛选粤港澳大湾区
df = df[df["Area0"].isin(['香港', '澳门', '广州市', '深圳市', '珠海市', '佛山市', '江门市', '中山市', '东莞市', '惠州市', '肇庆市'])]
```

```python
# 保存文件
df.to_file(r"E:\Netease\工程数据\粤港澳大湾区2.shp",encoding="utf-8")
```

### 制图

```python
request = cimgt.GoogleTiles()
```

```python
# 创建一个Stamen地形网络地图对象，设置缩放级别为8
stamen_terrain = cimgt.Stamen('terrain-background')
```

```python
zoom_level = 8
```

```python
# 创建地图画布
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.set_extent([111, 116, 21.5, 24.5], crs=ccrs.PlateCarree())  # 设置显示范围

# 添加地理特征
ax.add_feature(cartopy.feature.LAND)
ax.add_feature(cartopy.feature.OCEAN)
ax.add_feature(cartopy.feature.COASTLINE)
ax.add_feature(cartopy.feature.BORDERS, linestyle=':')
ax.add_feature(cartopy.feature.LAKES, alpha=0.5)
ax.add_feature(cartopy.feature.RIVERS)

# # 添加网络地图到地图上，设置透明度为0.5
ax.add_image(request, zoom_level, alpha=0.5)

# 读取粤港澳大湾区shp文件
shpfilename = r'E:\Netease\工程数据\粤港澳大湾区2.shp'
reader = shpreader.Reader(shpfilename)
records = reader.records()

# 绘制粤港澳大湾区边界
for record in records:
    ax.add_geometries([record.geometry], ccrs.PlateCarree(), facecolor='white', edgecolor='black',linestyle="--")  # 设置图例

# 添加城市名称和发展定位标签
city_names = ['香港', '澳门', '广州', '深圳', '珠海', '佛山', '江门', '中山', '东莞', '惠州', '肇庆']
city_lons = [114.16, 113.54, 113.26, 114.06, 113.57, 113.12, 113.08, 113.38, 113.75, 114.41, 112.46]
city_lats = [22.28, 22.19, 23.13, 22.54, 22.27, 23.02, 22.58, 22.52, 23.02, 23.11, 23.05]
city_roles = ['金融、航运、贸易中心',
              '旅游休闲中心',
              '商贸中心',
              '创新创意之都',
              '航空航天产业基地',
              '先进制造业基地',
              '侨乡经济发展示范区',
              '现代制造业基地',
              '现代服务业基地',
              '生态休闲旅游目的地',
              '生态屏障']
text_lst = []
for name, lon, lat, role in zip(city_names, city_lons, city_lats, city_roles):
    # 在城市位置添加点
    ax.plot(lon, lat, marker='o', color='red', transform=ccrs.Geodetic())
    # 在城市位置添加名称和发展定位
    text_lst.append(ax.text(lon, lat, name+": "+role,
            horizontalalignment='left', fontsize=8,
            transform=ccrs.Geodetic()))

# 调整标注位置
adjust_text(text_lst,)

# 添加网格线和刻度标签
ax.gridlines(draw_labels=True,linewidth=0.8,color='k', alpha=0.3, linestyle='--')

# 保存或显示地图
# plt.savefig('yuegangaodawanqu.png') # 可以保存为png或其他格式的图片文件
plt.show()  # 或者直接在屏幕上显示
```
