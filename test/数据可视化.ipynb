# 数据可视化

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titanic=sns.load_dataset('titanic')
```

## 分布情况

```python
sns.stripplot(x='embarked',y='age',hue='who',jitter=0.2,data=titanic)
```

```python
sns.violinplot(x='class',y='age',hue='who',data=titanic)
```

## 回归情况

```python
#导入数据集'iris'
iris=sns.load_dataset('iris')

#随机查看数据集的10行数据
iris.sample(10)
```

```python
sns.regplot(x='sepal_length',y='petal_length',data=iris)
```

```python
sns.lmplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```

## 相关热力图

```python
# 导入数据
flights = sns.load_dataset("flights")
flights.head(10)
```

```python
f=flights.pivot(index='year',columns='month',values='passengers')
cmap = sns.diverging_palette(200,20,sep=20,as_cmap=True)
sns.heatmap(f,cmap=cmap)
```
